```python
import pandas as pd

#import
movie_data = pd.read_csv('resources/movie_platforms.csv', low_memory=False)
movie_data.head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vector = TfidfVectorizer(stop_words='english')
movie_data['Title'] = movie_data['Title'].fillna('')
tfidf_matrix = tfidf_vector.fit_transform(movie_data['Title'])
```

```python
from sklearn.metrics.pairwise import linear_kernel
sim_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(movie_data.index, index=movie_data['Title']).drop_duplicates()
indices[:5]
```

```python
def movie_recommender(Title, sim_scores=sim_matrix):
    idx = indices[Title]
    sim_scores = list(enumerate(sim_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return movie_data['Title'].iloc[movie_indices]
```
